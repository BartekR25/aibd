```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

def function(x):
    return x ** 2 + 5
x_1 = np.linspace(-1, 1)
x_2 = np.linspace(-6, 6)
x_3 = np.linspace(0, 5)

plt.plot(x_1, function(x_1), label='$x^2 + 5$')
plt.legend(loc='upper center')
plt.xlabel("Argumenty funkcji")
plt.ylabel("Warto funkcji")
plt.title("Funkcja $x^2 + 5$ na przedziale (-1, 1)")
plt.grid(), plt.show()

plt.plot(x_2, function(x_2), label='$x^2 + 5$')
plt.legend(loc='upper center')
plt.xlabel("Argumenty funkcji")
plt.ylabel("Warto funkcji")
plt.title("Funkcja $x^2 + 5$ na przedziale (-6, 6)")
plt.grid(), plt.show()

plt.plot(x_3, function(x_3), label='$x^2 + 5$')
plt.legend(loc='upper left')
plt.xlabel("Argumenty funkcji")
plt.ylabel("Warto funkcji")
plt.title("Funkcja $x^2 + 5$ na przedziale (0, 5)")
plt.grid(), plt.show()
```

W powyszym wiczeniu stworzyem trzy wykresy dla funkcji x^2 + 5 dla rnych wartoci argumentw. Nazwaem je, podpisaem osie oraz dodaem legendy. Naoyem rwnie na wykresy siatk uatwiajc odczytywanie wartoci w poszczeglnych punktach.

```python
name_values = ['Ola', 'Ula', 'Jan', 'Kuba', 'Ala']
surname_values = ['Nowak', 'Mucha', 'Kowal', 'Kawka', 'Glik']
age_values = [22, 28, 29, 23, 25]
sex_values = ['K', 'K', 'M', 'M', 'K']
df = pd.DataFrame({'name': name_values, 'surname': surname_values, 'age': age_values,
                  'sex': sex_values})
print(df)
```

```python
df.info(verbose=True)
```

W zadaniu 4 stworzyem DataFrame o wskazanych kolumnach i uzupeniem je picioma przykadowymi rekordami. Nastpnie wywietliem informacj o danych uywajc df.info - moemy zobaczy jakie wielkoci z kolumn s jakim typem. df.describe nie dziaa na wartoci string, natomiast dziaa na wartoci typu int. Przykadowe wartoci zaprezentowane to warto maksymalna, minimalna czy rednia. df.head(n) pokazuje n pierwszych rekordw z DataFrame.
