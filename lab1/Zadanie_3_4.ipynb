```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

def function(x):
    return x ** 2 + 5
x_1 = np.linspace(-1, 1)
x_2 = np.linspace(-6, 6)
x_3 = np.linspace(0, 5)

plt.plot(x_1, function(x_1), label='$x^2 + 5$')
plt.legend(loc='upper center')
plt.xlabel("Argumenty funkcji")
plt.ylabel("Wartość funkcji")
plt.title("Funkcja $x^2 + 5$ na przedziale (-1, 1)")
plt.grid(), plt.show()

plt.plot(x_2, function(x_2), label='$x^2 + 5$')
plt.legend(loc='upper center')
plt.xlabel("Argumenty funkcji")
plt.ylabel("Wartość funkcji")
plt.title("Funkcja $x^2 + 5$ na przedziale (-6, 6)")
plt.grid(), plt.show()

plt.plot(x_3, function(x_3), label='$x^2 + 5$')
plt.legend(loc='upper left')
plt.xlabel("Argumenty funkcji")
plt.ylabel("Wartość funkcji")
plt.title("Funkcja $x^2 + 5$ na przedziale (0, 5)")
plt.grid(), plt.show()
```

W powyższym ćwiczeniu stworzyłem trzy wykresy dla funkcji x^2 + 5 dla różnych wartości argumentów. Nazwałem je, podpisałem osie oraz dodałem legendy. Nałożyłem również na wykresy siatkę ułatwiającą odczytywanie wartości w poszczególnych punktach.

```python
name_values = ['Ola', 'Ula', 'Jan', 'Kuba', 'Ala']
surname_values = ['Nowak', 'Mucha', 'Kowal', 'Kawka', 'Glik']
age_values = [22, 28, 29, 23, 25]
sex_values = ['K', 'K', 'M', 'M', 'K']
df = pd.DataFrame({'name': name_values, 'surname': surname_values, 'age': age_values,
                  'sex': sex_values})
print(df)
```
