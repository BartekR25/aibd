```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
DIR = "../AnalysisData/"
df = pd.read_csv(f"{DIR}13_WARMIN╠üSKO-MAZURSKIE.csv")
df = df.drop(df.columns[[0, 1]], axis=1)
df.to_csv(path_or_buf=f"{DIR}war_maz_melted.csv", index=False)
df
```

# Ilość poszczególnych ocen dla sklepu Beko

```python
df_beko = df[df['Marka'].isin(['Beko'])]
df_beko = df_beko.value_counts(['Ocena'], sort=False)
suma = df_beko.tolist()
df_beko.plot(x="Ocena", y=suma, kind="bar", figsize=(40,40), fontsize=40)
plt.ylabel('Ilość ocen')
```
